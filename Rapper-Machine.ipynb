Import library

```
import gpt_2_simple as gpt2
```

Download the GPT-2 Model

```
 gpt2.download_gpt2(model_name='355M')
```

```
# Load DATA
file_name = "/content/drive/MyDrive/l.txt.txt"
# Start session
sess = gpt2.start_tf_sess()
# Finetuning the model and start training
gpt2.finetune(sess, dataset=file_name, model_name='355M', steps=2000, restore_from='fresh', run_name='rap', print_every=10, sample_every=200, save_every=500) 
```

```
# Generate Rap
gpt2.generate(sess, temperature = 0.7, run_name='rap')
```
